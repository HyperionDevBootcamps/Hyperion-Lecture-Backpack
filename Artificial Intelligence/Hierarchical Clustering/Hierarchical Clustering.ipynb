## Types of Heirarchical Clustering Algorithms

#### Agglomerative

   - Start with each observation in its own cluster
   - Merge the two closest clusters
   - Continue until all objects are in a single cluster

   
#### Divisive

   - Start with all objects in a single cluster
   - Divide that cluster that will give the two furthest apart clusters
   - Continue until all objects are in their own cluster

## Profiling Clusters 

When we want to describe what makes our clusters unique, we look at the following:

 - Number of objects in a cluster
 - The objects in a cluster (a list of cluster members)
 - Average attributes of each cluster
