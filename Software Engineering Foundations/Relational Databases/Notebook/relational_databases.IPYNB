#### Relational Databases

- A relational database is a set of tables that store data to easily retrieve and manipulate data.
- They are widely used in many applications, such as facebook, google, e-commerce and many more.


##### Designing a Relational Databases

- Identify entities and their attributes.
    - An entity is a real-world object or a concept. This can be a customer or an order. 
    - An attribute is a portion of an entity that’s used to describe it, such as their name or the order ID.   
- Identify relationships between entities. An example would be customers place orders.
    - These are typically identified and implemented using foreign keys   
- Database design should be built using normalization form.
    - This is the process of organising data in different tables (Customers table and Order table)


##### Creating a Relational Database

- Once a schema (database) is designed the next step is to create the actual schema and it’s tables.
- Once the tables are created we can use SQL to insert and manipulate our statements.

##### Database Normilization

- Process in database design that helps organize data efficiently and reduces data redundancy.
- We break down a complex database into simpler, more manageable tables, while ensuring that data relationships are maintained.
- The main goal of normalization is to minimize data anomalies, such as insertion, update, and deletion anomalies. 

We have several levels of normalization, often referred to as normal forms. Each form with specific rules and requirements.

**First Normal Form (1NF):**

- Requires a primary key for each table.
- Ensures that all columns contain atomic (indivisible) values.
- Eliminates repeating groups of data within rows.

**Second Normal Form (2NF):**
- Must satisfy the requirements of 1NF.
- Non-key attributes (columns) should be functionally dependent on the entire primary key.
- Avoids partial dependencies, where non-key attributes depend on only part of the primary key.

**Third Normal Form (3NF):**
- Must satisfy the requirements of 1NF and 2NF.
- Eliminates transitive dependencies, where non-key attributes depend on other non-key attributes within the same table.
- Achieves a higher level of data independence and further reduces redundancy.
