## System Analysis and Design

#### Learning objectives

- System Analysis and Design
    - Define the diffirent types of requirements needed when designing a program.
    - Describe the requirements engineering process and use cases.
    - Explain system architecture and define some known architectures such as Layered, Repository, Client Server and Pipe and Filter.



##### Requirements

- Tells us what the system should be able to do
- If we have a program that organises files for us some requirements can be
    - Reading file names
    - Sort them in a certain order
    - Put them into specified folder
- Requirements can be divided into 2 categories
    - User requirements
    - System requirements


##### User Requirements

- What is important to the user
- Broad idea of what the program should be able to do
- Simple language anyone can understand


##### System Requirements

- Detailed version of the user requirements
- Technical jargon is used for developers to understand the exact functions and constraints
- System Requirements is further divided into 2 categories
    - Functional requirements
        - Activities required of the program
    - Non-functional requirements
        - Characteristics of the system as a whole


##### Functional Requirements

- Dependant on
    - The software being developed
    - The amount of users
    - The approach taken by the organization when writing the requirements


##### Non-Functional Requirements

- Not directly concerned with the specific activities that a system must perform or support
- Can sometimes be more critical than individual functional requirements
- It is not always easy to distinguish between functional and non-functional requirements so we use frameworks
- A popular framework that is used is FURPS
    - Functional requirements
    - Usability
    - Reliability
    - Performance
    - Security


- **Usability**
    - Describe operational characteristics related to users
        - User interface
        - Online help
        - Documentation
- **Reliability**
    - Dependability of the system
    - How the system detects and recovers from unexpected behaviour
- **Performance**
    - Describe operational characteristics related to measures of workload, such as throughput and response time
- **Security**
    - How access to program will be controlled
    - How data will be protected and secured




##### Requirements Document

- Once user and system requirements are determined, they are written up in an official requirements document
- Universal document that everyone from users and system developers will see
- Has to be general enough for the user to understand
- Detailed enough for developers to use a springboard for development
- User requirements will take the form of an introduction
- System requirements will be the body of the text

##### Requirements Engineering Process

- Process of finding, analysing, documenting, and checking the requirements of a system
- Requirements Engineers conduct a feasibility study to determine if the system is worth building
- Consists of analyzing the
    - Market
    - User requirements
    - System requirement  
- If it is decided that a system is worth building the engineers will follow a three step process  




- **Requirements Elicitation**
    - Determine who will use the system
    - Scenarios are discussed to get a better idea of what is required and how it will work
    - Generally where user requirements are determined  
- **Requirements Specification**
    - Write the requirements from Elicitation into a requirement document
    - Use cases are used to show how a user will interact with the system  
- **Requirements Validation**
    - Determine if documented requirements contain everything the user wants it to do
    - This is an important step to help prevent miscommunication  

##### Use Cases

- Activity the system performs usually in response to a request
- Identifies actors involved and names the type of interaction
- The set of use cases represents all the possible interactions with the system described in the requirements

  
  
  


##### Prototyping

- Basic mock-up of the system to see if development is heading in the right direction
- Is developed quickly and cost effectively
- Has to be developed enough for design experiments and to determine if requirements are being met
- Very important for UI design


##### UI Design Best Practices

- Keep it simple
- Use common UI elements consistently
- Use colours, textures and typography purposefully
- Communicate


##### UX vs UI

- User Experience vs User Interface
- UI deals specifically with the design elements of the interface
- UX is more abstract and deals with the feel of the program

##### System Architecture

- Blueprint based on your previously established requirements that will be the base of your designs for the system
- Aims to see
    - How the components of the system are organised and put together
    - How the components will communicate with each other
    - Be mindful of the constraints of the system as a whole
- Three advantages for implementing software architecture
    - Stakeholder communication
    - System analysis
    - Large-scale reuse


##### Layered Architecture

- All components are separated into layers
- Each layer can only access services from the layer below it
- Outer layers can be easily changed
- Only inner layers have to interface with the hardware or OS
  
- Most systems will have four layers
    - Presentation layer
        - User interface component
    - Business layer
        - Responsible for aspects such as user authentication and authorisation
    - Persistence layer
        - Deals with functional components
    - Database layer
        - Database component
  


##### Repository Architecture

- All data is kept in a single repository
- All components are separated but can interact through the repository
- Very useful if you have a lot of data
- Repository creates single point of failure
  


##### Client-Server Architecture

- Set of services developed to run on servers
- Clients access server through a network
- Network usage allows multiple clients to access the server and its services at the same time
- Network creates single point of failure
  


##### Pipe and Filter Architecture

- Data provided goes through a series of transformations(filters)
- Very useful pattern for data processing
- Because input has to be parsed and output unparsed computation time can increase making the system difficult to update or upgrade
