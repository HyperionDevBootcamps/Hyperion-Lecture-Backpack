   - Identifying and fitting visualisations

Data visualisation assists us with finding patterns and interesting bits of information within datasets that we analyse and to convey these findings with easy to communicate and follow visualisations.

Data scientists perform various functions to analyse data in meaningful ways, however, different people describe these steps in different ways. These steps do not follow a strict order of any sort, and can be describes as such: 
    
    (Data engineering)
        - Data Wrangling
        - Data Clensing
        - Data Preparation

    (Machine Learning)
        - Model Learning
        - Model Validation

    (Operations)
        - Model Deployment
        - Data Visualisation

Data Wrangling can be defined as :  "The process by which you identify, collect, merge, and preprocess one or more datasets in preparation for data clensing" (Jones 2018)

Once we've wrangled that data into a format which we can work with, we clean the data by fixing errors we come across within the dataset. This could potentially involve : 

    - Dealing with missing data
    - correcting formatting issues
    - fixing / removing incorrect data (Depending on severity)

The first job undertaken by a data scientist is to gain a good and clear understanding of the data that they are working with. The data scientist needs to understand what data they have, what format it is in, what format it needs to be converted to, what is wrong with the data and how to fix problems. Various visualisations (such as graphs, tables and charts) are used to gain better insight into the data. 

Therefore, data visualisation is important at both the end of the data science pipeline (presenting the findings of a study) and in te initial phase (data engineering phase) of the data science pipeline to help gain an understanding of data.

The Wine dataset is a classic dataset in machine learning and statistics. 

### Overview:

- Number of Instances: 178
- Number of Features: 13 numeric features which represent various chemical properties of wines
- Target Variable: Three classes of wines (class 0, class 1, and class 2)

### Features:

1. Alcohol: Alcohol content in the wine (measured in % v/v)
2. Malic Acid: Malic acid content in the wine (measured in g/l)
3. Ash: Ash content in the wine (measured in g/l)
4. Alcalinity of Ash: A measure of the alkalinity of the ash (measured in g/l)
5. Magnesium: Magnesium content in the wine (measured in mg/l)
6. Total Phenols: Total phenolic content in the wine (measured in mg/l)
7. Flavanoids: Flavanoid content in the wine (measured in mg/l)
8. Nonflavanoid Phenols: Nonflavanoid phenolic content in the wine (measured in mg/l)
9. Proanthocyanins: Proanthocyanidin content in the wine (measured in mg/l)
10. Color Intensity: Intensity of color in the wine (measured as absorbance at 420 nm)
11. Hue: Hue of the wine, representing the color (dimensionless)
12. OD280/OD315 of diluted wines:** Ratio of optical densities (unitless)
13. Proline: Proline content in the wine (measured in mg/l)

### Classes:

The target variable consists of three classes representing different types of wines:

- Class 0: Wines of the first type
- Class 1: Wines of the second type
- Class 2: Wines of the third type

### Dataset Source:

The Wine dataset is one of the built-in datasets in scikit-learn and was originally collected by Forina, M. et al. in their research work. It's often used for practicing classification algorithms and techniques in machine learning.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
```

```python
# Load Wine dataset from scikit-learn
wine = datasets.load_wine()
X, y = wine.data, wine.target

# Create a DataFrame 
df = pd.DataFrame(data=X, columns=wine.feature_names)

# Calculating the mean values for each feature
mean_values = df.mean()

# Plot a bar chart
plt.figure(figsize=(10, 6))
plt.bar(wine.feature_names,mean_values)
plt.xlabel('Features')
plt.ylabel('Mean Value')
plt.title('Mean Feature Values in Wine Dataset')
plt.xticks(rotation=45)
plt.show()
```

It's quite difficult to make comparisons across the different features because they are measured on different scales.

```python
# Standardize features for the bar plot
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

# Bar Chart with standardized features
plt.figure(figsize=(8, 6))
plt.bar(wine.feature_names, X_standardized.mean(axis=0))
plt.xlabel('Features')
plt.ylabel('Mean Value (Standardized)')
plt.title('Mean Standardized Feature Values in Wine Dataset')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(6, 6))
class_counts = [sum(y == i) for i in range(3)]
plt.pie(class_counts, labels=wine.target_names, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Wine Classes')
plt.show()
```

The bar chart is best used for data that is discrete and categorical
    

```python
'''
Line Graph - shows values over a time period. Multiple line graphs can show and compare many different
    categories over a time period.
        Typically, one axis shows data values, and the other axis depicts time.
'''

import matplotlib.pyplot as plt
import numpy as np

# make dummy data
x = [2009, 2010, 2011, 2012, 2013, 2014]
y = [12.4, 13.9, 11.6, 17.7, 17.1, 18.0]
z = [13.5, 13.7, 12.6, 15.9, 15.7, 17.5]

# create plots
fig, ax = plt.subplots()

# one can create multiple plots on one graph
ax.plot(x, y, color="r")
ax.plot(x,z)

# one can even name the axis with matplotlib
plt.title("INSERT TITLE HERE")
plt.xlabel("Year")
plt.ylabel("Profits")

plt.show()
```

The line graph is best used for data that is linear(continuous) or a time series.
    Keep in mind that time is usually on the x-axis

Working with Seaborn

        Seaborn is a data visualisation library that was built on top of matplotlib. While matplotlib is capable of providing basic graphs, such as line and scatter plots, seaborn can provide a bit more in terms of graphing. Additionally, it also integrates nicely with pandas.

For example : let's take data from a dataset for an individuals age and salary, we would like to understand the relationship between these two columns of data. 

Now, with matplotlib, the scatter() method would probably be best for taking a look at the data. 

```python
import matplotlib.pyplot as plt
import numpy as np

# dummy data
age = np.arange(50)
salary = np.random.uniform(20000, 78000, len(age))

plt.scatter(age, salary)
plt.show()
```

We could get a general sense of the data, however it may be difficult to understand the trends immediately. With seaborn, we can use a lineplot() method that automatically plots the averages and standard deviations for ease of reading.

This makes it a bit easier to see any trends or pattern within the data.

Some commonly use Seaborn plots would be :

        - histplot()
        - barplot()
        - boxplot()
