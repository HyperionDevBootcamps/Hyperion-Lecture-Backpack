## Introduction

Exploratory Data Analysis (EDA) involves systematically examining and summarizing the main characteristics, patterns and insights within a dataset

Why is it important?

1. Understanding data
2. Detecting patterns and trends
3. Feature selection
4. Hypothesis generation
5. EDA can help analysts make assumptions for modelling
6. Communication

## Descriptive Statistics

Before you dive into tackling a situation or problem, it is wise to step back and take a glimpse at the the big picture. We can use descriptive statistics to do so. We have two classes of descriptive statistics:

1. Measures of central tendency
2. Measures of spread

### Measures of Central Tendency

1. Mean or average: the sum of all values divided by the number of values. 




2. Median: the middle-most value in a sequence of numbers.



### Measures of Spread

1. Variance: an absolute measure of how spread out the data is.



#### Example: 

Lets say you are managing stock portfolios. An investor wants to find a low-risk stock in the tech industry. You consider two stocks:

 - Company A: last few stock prices were [128, 146, 112, 153] with mean 134.75
 - Company B: last few stock prices were [163, 52, 208, 128] with mean 137.75

While company B has a higher mean, we need to consider the variance of these prices since out investor is interested in low-risk stocks

 - Company A variance = 340.92
 - Company B variance = 4340.25

The decision is obvious now. Company A has stocks that tend to vary less from the mean.


2. The second measure of spread we will look at is Standard deviation: the square root of the variance. 

   Commonly used when defining outliers. When data points lie x standard deviations from the mean, they would be classified as an outlier.

### Why is identifying skewed data important?

We love to use the mean as a statistic of central tendancy. However, the mean gets heavily influenced by outliers. If you have a skewed dataset, the median is a better measure of central tendency it is not affected by extreme values.
